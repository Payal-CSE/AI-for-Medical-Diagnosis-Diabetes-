```python
#dIABETES DATASET CLEANING


# Import Libraries
import pandas as pd
import numpy as np


# Load Dataset
df = pd.read_csv("diabetes_data_not_cleaned.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)


# Rename Columns
df.columns = [
    "Age",
    "Gender",
    "Polyuria",
    "Polydipsia",
    "SuddenWeightLoss",
    "Weakness",
    "Polyphagia",
    "GenitalThrush",
    "VisualBlurring",
    "Itching",
    "Irritability",
    "DelayedHealing",
    "PartialParesis",
    "MuscleStiffness",
    "Alopecia",
    "Obesity",
    "Diabetes"
]


# Dataset Information
print("\nDataset Information")
print(df.info())

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()


# Clean Text Columns
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )


# Check Unique Values
print("\nUnique Values\n")

for col in text_columns:
    print(f"{col}: {df[col].unique()}")
    print()

# ------------------------------------------------------------
# Encode Yes / No
# ------------------------------------------------------------

yes_no_columns = [
    "Polyuria",
    "Polydipsia",
    "SuddenWeightLoss",
    "Weakness",
    "Polyphagia",
    "GenitalThrush",
    "VisualBlurring",
    "Itching",
    "Irritability",
    "DelayedHealing",
    "PartialParesis",
    "MuscleStiffness",
    "Alopecia",
    "Obesity"
]

for col in yes_no_columns:
    df[col] = df[col].replace({
        "Yes":1,
        "No":0
    })

# ------------------------------------------------------------
# Encode Gender
# ------------------------------------------------------------

df["Gender"] = df["Gender"].replace({
    "Male":1,
    "Female":0
})

# ------------------------------------------------------------
# Encode Target
# ------------------------------------------------------------

df["Diabetes"] = df["Diabetes"].replace({
    "Positive":1,
    "Negative":0
})

# ------------------------------------------------------------
# Convert Encoded Columns to Integer
# ------------------------------------------------------------

columns_to_convert = yes_no_columns + ["Gender", "Diabetes"]

for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# ------------------------------------------------------------
# Final Checks
# ------------------------------------------------------------

print("\nFinal Missing Values")
print(df.isnull().sum())

print("\nData Types")
print(df.dtypes)

print("\nFirst Five Rows")
print(df.head())

print("\nShape:", df.shape)

# ------------------------------------------------------------
# Save Clean Dataset
# ------------------------------------------------------------

df.to_csv("diabetes_cleaned.csv", index=False)

print("\nCleaning Completed Successfully!")
```
